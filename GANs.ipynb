# 1. What does GAN stand for, and what is its main purpose?


**GAN** stands for **Generative Adversarial Network**.

**Main Purpose:**

* GANs are a type of deep learning model designed to **generate new, realistic data** (images, audio, text, etc.) that resemble a given training dataset.
* They consist of **two networks**:

  1. **Generator:** Creates fake data.
  2. **Discriminator:** Tries to distinguish real data from fake data.

The generator and discriminator **compete in a game-like setup**, improving each other until the generator produces highly realistic data.

**In short:** GANs are used for **data generation, synthesis, and augmentation** by learning to mimic real data distributions.


# 2. Explain the concept of the "discriminator" in GANs?


**Discriminator in GANs:**

The **discriminator** is a neural network in a GAN that **distinguishes real data from fake data** generated by the generator.

* It outputs a **probability** indicating whether an input sample is real (from the training dataset) or fake (from the generator).
* During training, the discriminator **learns to correctly classify real vs. fake**, while the generator tries to **fool the discriminator**.

**In short:** The discriminator acts as a **critic** that guides the generator to produce more realistic data.

# 3. How does a GAN work?


**How a GAN Works:**

A **Generative Adversarial Network (GAN)** consists of two neural networks, a **Generator** and a **Discriminator**, that compete with each other:

1. **Generator (G):** Takes random noise as input and generates fake data resembling the real dataset.
2. **Discriminator (D):** Evaluates whether an input sample is **real** (from the dataset) or **fake** (from the generator).
3. **Adversarial Training:**

   * The discriminator learns to **distinguish real from fake**.
   * The generator learns to **fool the discriminator** by producing more realistic data.

This iterative process continues until the generator creates data that the discriminator **cannot reliably tell apart from real data**.

**In short:** GANs work as a **game between two networks**, improving the generator’s ability to create realistic data over time.


# 4. What is the generator's role in a GAN?


**Generator in a GAN:**

The **generator** is a neural network in a GAN that **creates new data samples** from random noise.

* Its goal is to **produce data that resembles the real dataset**.
* During training, it tries to **fool the discriminator** into thinking its generated samples are real.

**In short:** The generator is the **creative component** of a GAN, responsible for **generating realistic synthetic data**.


# 5. What is the loss function used in the training of GANs?


**Loss Function in GANs:**

GANs use a **minimax loss function** that represents the adversarial game between the **generator (G)** and **discriminator (D)**:


$$
\min_G \max_D V(D, G) = \mathbb{E}_{x \sim p_{\text{data}}}[\log D(x)] + \mathbb{E}_{z \sim p_z}[\log(1 - D(G(z)))]
$$


* **Discriminator loss:** Encourages D to assign high probability to real data and low probability to fake data.
* **Generator loss:** Encourages G to generate data that maximizes D’s probability of being real (i.e., fool the discriminator).

**In short:** The loss function makes the generator and discriminator **compete**, improving G’s realism and D’s discrimination ability over time.

# 6. What is the difference between a WGAN and a traditional GAN?



**Difference Between WGAN and Traditional GAN:**

| Aspect                   | Traditional GAN                         | WGAN (Wasserstein GAN)                                 |
| ------------------------ | --------------------------------------- | ------------------------------------------------------ |
| **Loss Function**        | Minimax loss with log probabilities     | Wasserstein (Earth-Mover) distance                     |
| **Training Stability**   | Can be unstable, prone to mode collapse | More stable, smoother training                         |
| **Discriminator Output** | Probability (0–1)                       | Real-valued score (critic), no sigmoid                 |
| **Gradient Behavior**    | Can vanish if D is too strong           | Provides meaningful gradients even if critic is strong |
| **Weight Constraint**    | No constraints                          | Requires weight clipping or gradient penalty           |

**In short:** WGAN replaces the traditional GAN’s discriminator with a **critic that estimates Wasserstein distance**, leading to **more stable training and better gradient behavior**.

# 7. How does the training of the generator differ from that of the discriminator?


**Difference Between Training the Generator and the Discriminator in GANs:**

* **Discriminator Training:**
  The discriminator is trained to **correctly classify real data as real and generated data as fake**. Its weights are updated to **maximize classification accuracy** between real and fake samples.

* **Generator Training:**
  The generator is trained to **produce realistic data** that can **fool the discriminator**. Its weights are updated to **minimize the discriminator’s ability to distinguish fake data from real data**.

**In short:** The discriminator learns to **detect fake data**, while the generator learns to **create fake data that looks real** through adversarial training.

# 8. What is a DCGAN, and how is it different from a traditional GAN?



**DCGAN (Deep Convolutional GAN):**

A **DCGAN** is a type of GAN that uses **deep convolutional neural networks** instead of fully connected layers in both the generator and discriminator.

**How DCGAN differs from a traditional GAN:**

* **Architecture:** DCGAN uses **convolutional layers** (CNNs), while traditional GANs mainly use fully connected layers.
* **Feature Learning:** Convolutions allow DCGANs to **learn spatial features**, making them well-suited for images.
* **Stability:** DCGAN applies architectural guidelines (no pooling layers, batch normalization), resulting in **more stable training**.
* **Image Quality:** Produces **higher-quality and more realistic images** compared to basic GANs.

**In short:** DCGAN improves GAN performance for images by leveraging **convolutional networks**, leading to **better stability and realism**.

# 9. Explain the concept of "controllable generation" in the context of GAN?



**Controllable Generation in GANs:**

**Controllable generation** refers to the ability to **guide or control specific attributes** of the data generated by a GAN, rather than producing completely random outputs.

This is achieved by:

* **Conditioning the GAN** on additional information such as class labels, attributes, or text descriptions (e.g., Conditional GANs).
* **Latent space manipulation**, where changing certain input variables alters specific features in the generated output (like color, shape, or style).

**Examples:**

* Generating images of a specific class (dogs, cars, faces).
* Controlling facial attributes such as age, expression, or hair color.

**In short:** Controllable generation allows GANs to **produce customized outputs** by steering the generation process toward desired characteristics.


# 10. What is the primary goal of training a GAN?

**Primary Goal of Training a GAN:**

The primary goal of training a **Generative Adversarial Network (GAN)** is to **learn the underlying data distribution** of the training dataset so that the generator can produce **new, realistic samples** that are indistinguishable from real data.

This is achieved through **adversarial training**, where:

* The **generator** learns to create realistic data.
* The **discriminator** learns to distinguish real data from generated data.

**In short:** The goal of a GAN is to **generate high-quality, realistic data** by learning from real examples.


# 11. What are the limitations of GANs?


**Limitations of GANs:**

1. **Training Instability:** GANs are difficult to train and may fail to converge due to the adversarial setup.
2. **Mode Collapse:** The generator may produce limited varieties of outputs, ignoring parts of the data distribution.
3. **Evaluation Difficulty:** There is no simple, universal metric to evaluate the quality of generated samples.
4. **High Computational Cost:** Training GANs often requires large datasets and powerful hardware (GPUs).
5. **Sensitivity to Hyperparameters:** Performance strongly depends on learning rates, network architecture, and loss functions.

**In short:** GANs can generate realistic data but suffer from **unstable training, mode collapse, and evaluation challenges**.


# 12. What are StyleGANs, and what makes them unique?


**StyleGANs (Style-based Generative Adversarial Networks)** are an advanced type of GAN designed to generate high-quality and realistic images. Unlike traditional GANs, StyleGANs use a style-based generator architecture that gives control over different visual features of an image at various levels.

StyleGANs introduce a **mapping network** that transforms a random noise vector into an intermediate latent space, which is then used to control the generator. This allows different layers of the generator to manage different aspects of the image. Early layers control coarse features such as pose and shape, middle layers control facial features like eyes and nose, and later layers control fine details such as skin texture and hair.

A key feature of StyleGANs is **Adaptive Instance Normalization (AdaIN)**, which injects style information into each layer of the generator. This enables precise and independent control over image attributes. StyleGANs also add **stochastic noise** at different layers to introduce natural variations, improving realism.

Another unique property is **disentangled feature control**, which allows individual attributes to be modified without affecting others. StyleGANs support **style mixing**, where styles from different latent codes can be combined in a single image.

Overall, StyleGANs are unique because they provide fine-grained control over image generation while producing extremely realistic and high-resolution images.


# 13. What is the role of noise in a GAN?


**Role of Noise in a GAN**

In a Generative Adversarial Network (GAN), noise plays a crucial role in enabling the generator to produce diverse and realistic outputs. The noise is usually a random vector sampled from a simple distribution such as a normal or uniform distribution and is provided as input to the generator.

The noise vector serves as a source of randomness that allows the generator to create different outputs each time, even when trained on the same dataset. By learning to transform this noise into meaningful data, the generator maps points from the noise space to the data space.

Noise helps prevent the generator from producing identical outputs, thereby reducing the problem of mode collapse. It also enables smooth interpolation between generated samples, meaning small changes in the noise vector lead to gradual changes in the generated output.

Overall, noise is essential because it introduces variability, supports diversity in generated samples, and allows the generator to learn the underlying data distribution effectively.


# 14. How does the loss function in a WGAN improve training stability?


In a Wasserstein GAN (WGAN), the loss function is designed to measure the distance between the real data distribution and the generated data distribution using the **Wasserstein (Earth Mover’s) distance** instead of the Jensen–Shannon divergence used in traditional GANs.

The Wasserstein distance provides **smooth and meaningful gradients** even when the real and generated distributions do not overlap. This allows the generator to continue learning throughout training, avoiding the vanishing gradient problem commonly seen in standard GANs.

In WGAN, the discriminator is replaced by a **critic** that outputs a real-valued score rather than a probability. The loss reflects how well the critic can distinguish real samples from fake ones based on this score, leading to more stable updates.

Additionally, WGAN enforces a **Lipschitz constraint** (using weight clipping or gradient penalty), which further stabilizes training by preventing extreme parameter updates.

Overall, the WGAN loss improves training stability by providing reliable gradients, reducing mode collapse, and ensuring smoother convergence during GAN training.

# 15. Describe the architecture of a typical GAN?

**Architecture of a Typical GAN**

A typical Generative Adversarial Network (GAN) consists of two main neural networks: the **Generator** and the **Discriminator**, which are trained simultaneously in an adversarial manner.

The **Generator (G)** takes a random noise vector sampled from a simple distribution (such as Gaussian or uniform) as input and transforms it into synthetic data that resembles real data. Its goal is to learn the underlying data distribution and produce realistic samples that can fool the discriminator.

The **Discriminator (D)** receives either real data from the training dataset or fake data generated by the generator. It acts as a binary classifier and outputs a probability indicating whether the input data is real or fake.

During training, the discriminator is optimized to correctly distinguish real samples from generated ones, while the generator is optimized to produce samples that the discriminator classifies as real. This creates a competitive process where both networks improve over time.

Both the generator and discriminator are usually implemented using deep neural networks such as fully connected networks or convolutional neural networks, depending on the data type.

Overall, a GAN architecture relies on this adversarial setup to learn and generate realistic data.

# 16. What challenges do GANs face during training, and how can they be addressed?


**Challenges in Training GANs and How They Can Be Addressed**

Training Generative Adversarial Networks (GANs) is challenging due to the adversarial nature of the generator and discriminator. One major issue is **training instability**, where the generator and discriminator fail to converge properly. This can be addressed by using improved architectures, careful learning rate tuning, and training the discriminator and generator in a balanced manner.

Another common problem is **mode collapse**, where the generator produces limited varieties of outputs. This can be reduced by techniques such as minibatch discrimination, feature matching, using Wasserstein GANs (WGANs), or adding noise and regularization.

**Vanishing gradients** can occur when the discriminator becomes too strong, causing the generator to stop learning. This issue can be addressed by using alternative loss functions like the Wasserstein loss or by limiting the discriminator’s capacity.

GANs are also sensitive to **hyperparameter selection**, including learning rates, batch size, and network depth. Careful tuning, normalization techniques (such as batch normalization), and gradient penalties can help stabilize training.

Finally, GAN training can suffer from **poor evaluation metrics**, making it difficult to assess progress. Using metrics like FID (Fréchet Inception Distance) and Inception Score helps evaluate generated samples more reliably.

Overall, these challenges can be addressed through improved loss functions, architectural changes, regularization techniques, and careful training strategies.


# 17. How does DCGAN help improve image generation in GANs?



**How DCGAN Improves Image Generation in GANs**

Deep Convolutional Generative Adversarial Networks (DCGANs) improve image generation by replacing fully connected layers with **convolutional neural networks (CNNs)** in both the generator and discriminator. This allows the model to better capture spatial features and local patterns in images.

In DCGAN, the **generator uses transposed convolution (deconvolution) layers** to progressively upsample the noise vector into a full-sized image. This helps generate structured and detailed images. The **discriminator uses convolutional layers** to efficiently extract hierarchical features and distinguish real images from fake ones.

DCGANs apply **batch normalization** to stabilize training and accelerate convergence, while **removing pooling layers** and using strided convolutions for downsampling. This design improves gradient flow and reduces training instability.

By leveraging convolutional architectures, DCGANs produce **sharper, more realistic images**, reduce artifacts, and provide more stable training compared to traditional GANs.

# 18. What are the key differences between a traditional GAN and a StyleGAN?


**Key Differences Between a Traditional GAN and a StyleGAN**

A traditional GAN uses a simple architecture where a random noise vector is directly fed into the generator to produce an output image. The generator learns to create images as a whole, with limited control over individual features. Changes in the input noise usually affect the entire image in an entangled manner.

In contrast, StyleGAN introduces a **style-based generator architecture**. Instead of directly using the noise vector, StyleGAN first maps it into an intermediate latent space through a **mapping network**. This intermediate representation controls different layers of the generator, allowing separate control over coarse, medium, and fine image features.

Traditional GANs offer limited control over image attributes and often suffer from entangled feature representations. StyleGANs provide **disentangled feature control**, making it possible to independently modify attributes such as pose, shape, and texture.

Additionally, StyleGAN injects **stochastic noise** at multiple layers to create realistic variations and uses techniques like **Adaptive Instance Normalization (AdaIN)** for style control. As a result, StyleGANs generate higher-quality, more realistic images and enable controllable image synthesis compared to traditional GANs.

# 19. How does the discriminator decide whether an image is real or fake in a GAN?


**How the Discriminator Decides Whether an Image Is Real or Fake in a GAN**

In a Generative Adversarial Network (GAN), the discriminator is a neural network that acts as a binary classifier. It receives an image as input, which can be either a real image from the training dataset or a fake image generated by the generator.

The discriminator processes the input image through its layers to extract features and then outputs a single value representing the probability that the image is real. A value close to 1 indicates that the image is considered real, while a value close to 0 indicates that it is considered fake.

During training, the discriminator learns by comparing its predictions with the true labels (real or fake) and minimizing a classification loss. As training progresses, the discriminator becomes better at identifying subtle differences between real and generated images.

At the same time, the generator is trained to produce images that can fool the discriminator. This adversarial process improves both networks, leading to more realistic generated images.

# 22. What is Mode Collapse in GANs, and how can it be prevented?


**Mode Collapse in GANs and How to Prevent It**

**Mode Collapse** is a common problem in GANs where the generator produces a limited variety of outputs, often generating very similar or identical samples regardless of the input noise. This happens when the generator finds a few outputs that consistently fool the discriminator and keeps producing them, ignoring other possible variations in the data distribution.

**Ways to Prevent Mode Collapse:**

1. **Use Improved GAN Architectures**: Techniques like Wasserstein GANs (WGANs) provide smoother gradients and reduce mode collapse.
2. **Mini-batch Discrimination**: Introduce features that allow the discriminator to look at multiple samples together, encouraging diversity.
3. **Feature Matching**: Train the generator to match statistics of real data features rather than only trying to fool the discriminator.
4. **Noise Injection**: Adding small noise at different layers of the generator can promote variation in outputs.
5. **Regularization and Training Tricks**: Techniques like gradient penalty, label smoothing, or careful balancing of generator and discriminator updates help maintain diversity.

By applying these methods, the generator can produce a wider variety of outputs and better capture the full data distribution.

# 1. Implement a simple GAN architecture to generate random images (like noise or basic shapes) using TensorFlow/Keras?

```python
# Import libraries
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
tf.random.set_seed(42)
np.random.seed(42)

# ---------------------
# Generator
# ---------------------
def build_generator(latent_dim=100):
    model = models.Sequential()
    model.add(layers.Dense(128, activation='relu', input_dim=latent_dim))
    model.add(layers.Dense(256, activation='relu'))
    model.add(layers.Dense(512, activation='relu'))
    model.add(layers.Dense(28*28, activation='tanh'))  # Output shape for 28x28 image
    model.add(layers.Reshape((28,28,1)))
    return model

# ---------------------
# Discriminator
# ---------------------
def build_discriminator(img_shape=(28,28,1)):
    model = models.Sequential()
    model.add(layers.Flatten(input_shape=img_shape))
    model.add(layers.Dense(512, activation='relu'))
    model.add(layers.Dense(256, activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))  # Output: probability real/fake
    return model

# ---------------------
# Build GAN
# ---------------------
latent_dim = 100
generator = build_generator(latent_dim)
discriminator = build_discriminator()
discriminator.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Combined GAN (Generator + Discriminator)
discriminator.trainable = False
gan_input = layers.Input(shape=(latent_dim,))
gan_output = discriminator(generator(gan_input))
gan = models.Model(gan_input, gan_output)
gan.compile(loss='binary_crossentropy', optimizer='adam')

# ---------------------
# Training
# ---------------------
def train_gan(gan, generator, discriminator, epochs=1000, batch_size=32):
    # Create random real images (like simple shapes)
    # Here we use random noise as "real" data
    X_real = np.random.rand(1000,28,28,1) * 2 - 1  # Normalize to [-1,1]
    y_real = np.ones((batch_size,1))
    y_fake = np.zeros((batch_size,1))

    for epoch in range(epochs):
        # -----------------
        # Train Discriminator
        # -----------------
        idx = np.random.randint(0, X_real.shape[0], batch_size)
        real_imgs = X_real[idx]

        noise = np.random.normal(0,1,(batch_size, latent_dim))
        fake_imgs = generator.predict(noise)

        d_loss_real = discriminator.train_on_batch(real_imgs, y_real)
        d_loss_fake = discriminator.train_on_batch(fake_imgs, y_fake)
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

        # -----------------
        # Train Generator
        # -----------------
        noise = np.random.normal(0,1,(batch_size, latent_dim))
        g_loss = gan.train_on_batch(noise, y_real)  # Try to fool discriminator

        # Print progress
        if epoch % 100 == 0:
            print(f"Epoch {epoch} | D Loss: {d_loss[0]:.4f}, D Acc: {d_loss[1]*100:.2f}% | G Loss: {g_loss:.4f}")

# ---------------------
# Train the GAN
# ---------------------
train_gan(gan, generator, discriminator, epochs=1000, batch_size=32)

# ---------------------
# Generate Images
# ---------------------
def generate_images(generator, latent_dim=100, n_images=5):
    noise = np.random.normal(0,1,(n_images, latent_dim))
    gen_imgs = generator.predict(noise)
    gen_imgs = 0.5 * gen_imgs + 0.5  # Rescale to [0,1]

    plt.figure(figsize=(10,2))
    for i in range(n_images):
        plt.subplot(1,n_images,i+1)
        plt.imshow(gen_imgs[i,:,:,0], cmap='gray')
        plt.axis('off')
    plt.show()

generate_images(generator)
```

# 4. Implement WGAN by modifying the loss function in the GAN?

```python
import torch
import torch.nn as nn
import torch.optim as optim

# ===== Step 1: Define Generator =====
class Generator(nn.Module):
    def __init__(self, noise_dim=100, data_dim=1):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(noise_dim, 64),
            nn.ReLU(),
            nn.Linear(64, data_dim)
        )

    def forward(self, z):
        return self.model(z)

# ===== Step 2: Define Critic (NOT Discriminator) =====
class Critic(nn.Module):
    def __init__(self, data_dim=1):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(data_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 1)   # No sigmoid in WGAN
        )

    def forward(self, x):
        return self.model(x)

# ===== Step 3: Initialize models =====
noise_dim = 100
G = Generator(noise_dim)
C = Critic()

# ===== Step 4: Optimizers =====
optimizer_G = optim.RMSprop(G.parameters(), lr=0.00005)
optimizer_C = optim.RMSprop(C.parameters(), lr=0.00005)

# ===== Step 5: Training loop (WGAN loss) =====
for step in range(100):
    # ----- Train Critic -----
    for _ in range(5):  # Critic trained more than Generator
        real_data = torch.randn(32, 1)   # Real samples
        z = torch.randn(32, noise_dim)
        fake_data = G(z).detach()

        loss_critic = -(torch.mean(C(real_data)) - torch.mean(C(fake_data)))

        optimizer_C.zero_grad()
        loss_critic.backward()
        optimizer_C.step()

        # Weight clipping (WGAN requirement)
        for p in C.parameters():
            p.data.clamp_(-0.01, 0.01)

    # ----- Train Generator -----
    z = torch.randn(32, noise_dim)
    fake_data = G(z)
    loss_generator = -torch.mean(C(fake_data))

    optimizer_G.zero_grad()
    loss_generator.backward()
    optimizer_G.step()

    if step % 10 == 0:
        print(f"Step {step} | Critic Loss: {loss_critic.item():.4f} | Generator Loss: {loss_generator.item():.4f}")
```

# 6. Create a StyleGAN-inspired architecture that outputs high-resolution images?

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt

# ===== Mapping Network (Z → W) =====
class MappingNetwork(nn.Module):
    def __init__(self, z_dim=512, w_dim=512):
        super().__init__()
        self.mapping = nn.Sequential(
            nn.Linear(z_dim, w_dim),
            nn.ReLU(),
            nn.Linear(w_dim, w_dim),
            nn.ReLU()
        )

    def forward(self, z):
        return self.mapping(z)

# ===== Style Block =====
class StyleBlock(nn.Module):
    def __init__(self, in_channels, out_channels, w_dim):
        super().__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, 3, padding=1)
        self.style = nn.Linear(w_dim, out_channels * 2)

    def forward(self, x, w):
        style = self.style(w)
        scale, bias = style.chunk(2, dim=1)
        scale = scale.unsqueeze(-1).unsqueeze(-1)
        bias = bias.unsqueeze(-1).unsqueeze(-1)

        x = self.conv(x)
        x = scale * x + bias
        return F.leaky_relu(x, 0.2)

# ===== StyleGAN-Inspired Generator =====
class StyleGANGenerator(nn.Module):
    def __init__(self, z_dim=512, w_dim=512):
        super().__init__()
        self.mapping = MappingNetwork(z_dim, w_dim)

        self.const_input = nn.Parameter(torch.randn(1, 512, 4, 4))

        self.block1 = StyleBlock(512, 256, w_dim)
        self.block2 = StyleBlock(256, 128, w_dim)
        self.block3 = StyleBlock(128, 64, w_dim)

        self.to_rgb = nn.Conv2d(64, 3, 1)

    def forward(self, z):
        w = self.mapping(z)
        x = self.const_input.repeat(z.size(0), 1, 1, 1)

        x = self.block1(x, w)
        x = F.interpolate(x, scale_factor=2)

        x = self.block2(x, w)
        x = F.interpolate(x, scale_factor=2)

        x = self.block3(x, w)
        x = F.interpolate(x, scale_factor=2)

        return torch.tanh(self.to_rgb(x))

# ===== Generate and Display Images =====
generator = StyleGANGenerator()
generator.eval()

z = torch.randn(8, 512)
images = generator(z)

plt.figure(figsize=(8,4))
for i in range(8):
    plt.subplot(2, 4, i+1)
    img = (images[i].permute(1,2,0).detach().numpy() + 1) / 2
    plt.imshow(img)
    plt.axis("off")

plt.suptitle("StyleGAN-Inspired Generated Images (64×64)")
plt.show()
```

# 7. Implement the Wasserstein loss function for GAN training?

```python
import torch

# ===== Wasserstein Loss Functions =====

def wasserstein_critic_loss(real_scores, fake_scores):
    """
    Wasserstein loss for the Critic (Discriminator).

    Args:
        real_scores (Tensor): Critic scores for real data
        fake_scores (Tensor): Critic scores for fake data

    Returns:
        Tensor: Critic loss
    """
    return -(torch.mean(real_scores) - torch.mean(fake_scores))


def wasserstein_generator_loss(fake_scores):
    """
    Wasserstein loss for the Generator.

    Args:
        fake_scores (Tensor): Critic scores for fake data

    Returns:
        Tensor: Generator loss
    """
    return -torch.mean(fake_scores)

# ===== Example Usage =====
# Simulated critic outputs
real_scores = torch.randn(32, 1)   # Critic output for real samples
fake_scores = torch.randn(32, 1)   # Critic output for fake samples

critic_loss = wasserstein_critic_loss(real_scores, fake_scores)
generator_loss = wasserstein_generator_loss(fake_scores)

print("Critic Loss:", critic_loss.item())
print("Generator Loss:", generator_loss.item())
```
